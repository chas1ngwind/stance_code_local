```python
import pandas as pd
```

```python
df_oppsite=pd.read_csv("D:/Jupyter/stance_code_local/failed_samples/100_gold1/test.tsv",sep="\t", header=None, index_col=False)
new_score= pd.read_csv('D:/Jupyter/stance_code_local/evaluation/bert_local_output/100_gold1/opposite_raw_score.csv',)
df_small= pd.read_excel('D:/Jupyter/stance_code_local/failed_samples/100_opposite_gold_1.xlsx', index_col=0)
```

```python
new_score
```

```python
df_small.rename(columns={'undermine_score':'undermine_origin','support_score':'support_origin', 'predict_label':'predict_origin', 'gold_label':'gold_origin'},inplace=True) 
```

```python
new_score.rename(columns={'undermine_score':'undermine_opposite','support_score':'support_opposite', 'predict_label':'predict_opposite', 'gold':'gold_opposite' },inplace=True) 
```

```python
integration = pd.concat([df_small.reset_index(drop=True), new_score], axis = 1)
```

```python
integration
```

```python
integration["oppose_sum"]=integration.undermine_origin+integration.support_opposite
integration["support_sum"]=integration.support_origin+integration.undermine_opposite
```

```python
# integration.drop('predict_label',axis=1,inplace=True)
```

```python
integration.loc[integration['oppose_sum']==integration['support_sum'], 'final_label'] = '0'
integration.loc[integration['oppose_sum']>integration['support_sum'], 'final_label'] = '0'
integration.loc[integration['oppose_sum']<integration['support_sum'], 'final_label'] = '1'
```

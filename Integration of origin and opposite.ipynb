### Grid Search

```python
####   grid search start
best_score = 0
for undermine_origin in (list(np.arange(0,2.5,0.25))+[10,100,1000]):
    for support_opposite in (list(np.arange(0,2.5,0.25))+[10,100,1000]):
        for support_origin in (list(np.arange(0,2.5,0.25))+[10,100,1000]):
            for undermine_opposite in (list(np.arange(0,2.5,0.25))+[10,100,1000]):
                integration["oppose_sum"]=undermine_origin*integration.undermine_origin+support_opposite*integration.support_opposite
                integration["support_sum"]=support_origin*integration.support_origin+undermine_opposite*integration.undermine_opposite
                integration.loc[integration['oppose_sum']==integration['support_sum'], 'final_label'] = 0
                integration.loc[integration['oppose_sum']>integration['support_sum'], 'final_label'] = 0
                integration.loc[integration['oppose_sum']<integration['support_sum'], 'final_label'] = 1
                score = len(integration[integration["gold_origin"] == integration["final_label"]])
                        

                if score > best_score:
                    best_score = score
                    best_parameters = {'undermine_origin':undermine_origin,'support_opposite':support_opposite, 'support_origin':support_origin, 'undermine_opposite':undermine_opposite}
####   grid search end
print("Best score:{:.2f}".format(best_score))
print("Best parameters:{}".format(best_parameters))
```

```python
df_small.rename(columns={'undermine_score':'undermine_origin','support_score':'support_origin', 'predict_label':'predict_origin', 'gold_label':'gold_origin'},inplace=True) 
```

```python
new_score.rename(columns={'undermine_score':'undermine_opposite','support_score':'support_opposite', 'predict_label':'predict_opposite', 'gold':'gold_opposite' },inplace=True) 
```

```python
integration = pd.concat([df_small.reset_index(drop=True), new_score], axis = 1)
```

```python
integration
```

```python
integration["oppose_sum"]=1.5*integration.undermine_origin+0.5*integration.support_opposite
integration["support_sum"]=1.5*integration.support_origin+0.5*integration.undermine_opposite
```

```python
# integration.drop('predict_label',axis=1,inplace=True)
```

```python
integration.loc[integration['oppose_sum']==integration['support_sum'], 'final_label'] = 0
integration.loc[integration['oppose_sum']>integration['support_sum'], 'final_label'] = 0
integration.loc[integration['oppose_sum']<integration['support_sum'], 'final_label'] = 1
```

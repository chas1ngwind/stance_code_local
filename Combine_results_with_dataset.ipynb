```python
import pandas as pd
```

```python
df_fail_results = pd.read_csv("../stance_code/Evaluation/antonym_output/antonym_pers_100_failed_gold0_stancy_bs32_lr1e_5_raw_score.csv")
df_success_results = pd.read_csv("../stance_code/Evaluation/antonym_output/antonym_pers_100_successed_gold0_stancy_bs32_lr1e_5_raw_score.csv")
```

```python
df_fail_results.rename(columns={'undermine_score':'undermine_opposite','support_score':'support_opposite', 'predict_label':'predict_opposite', 'gold':'gold_opposite' },inplace=True)
df_success_results.rename(columns={'undermine_score':'undermine_opposite','support_score':'support_opposite', 'predict_label':'predict_opposite', 'gold':'gold_opposite' },inplace=True)
```

```python
df_fail_results
```

```python
df_fail_ori = pd.read_csv("../stance_code/Dataset/manul_opposite/stancy_100_gold0/failed/test.tsv", delimiter="\t")
df_success_ori = pd.read_csv("../stance_code/Dataset/manul_opposite/stancy_100_gold0/succeeded/test.tsv", delimiter="\t")
```

```python
df_fail_ori.rename(columns={'undermine_score':'undermine_origin','support_score':'support_origin', 'predict_label':'predict_origin', 'gold':'gold_origin'},inplace=True)
df_success_ori.rename(columns={'undermine_score':'undermine_origin','support_score':'support_origin', 'predict_label':'predict_origin', 'gold':'gold_origin'},inplace=True)
```

```python
df_success_ori
```

```python
integration = pd.concat([df_fail_ori, df_fail_results],axis=1)
integration_suc = pd.concat([df_success_ori, df_success_results],axis=1)
```

```python
integration
```

```python
integration["oppose_sum"]=integration.undermine_origin-integration.undermine_opposite
integration["support_sum"]=integration.support_origin-integration.support_opposite
```

```python
integration_suc["oppose_sum"]=integration_suc.undermine_origin-integration_suc.undermine_opposite
integration_suc["support_sum"]=integration_suc.support_origin-integration_suc.support_opposite
```

```python
integration
```

```python
integration.loc[integration['oppose_sum']==integration['support_sum'], 'final_label'] = 0
integration.loc[integration['oppose_sum']>integration['support_sum'], 'final_label'] = 0
integration.loc[integration['oppose_sum']<integration['support_sum'], 'final_label'] = 1
```

```python
integration_suc.loc[integration_suc['oppose_sum']==integration_suc['support_sum'], 'final_label'] = 0
integration_suc.loc[integration_suc['oppose_sum']>integration_suc['support_sum'], 'final_label'] = 0
integration_suc.loc[integration_suc['oppose_sum']<integration_suc['support_sum'], 'final_label'] = 1
```
